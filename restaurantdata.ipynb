# 1. Data Understanding and Cleaning:

```python
dfSwiggy=pd.read_csv("D:\Personal\Training\Project-Swiggy\swiggy.csv",na_values=["--","NA"]);
dfSwiggy.head()
```

```python
dfSwiggy.drop(["lic_no","menu"], axis=1, inplace=True)
dfSwiggy.info()
dfSwiggy.describe()
```

```python
from random import randint

def convert_ratings(rating):
    if rating == 'Too Few Ratings':
        return randint(0, 20)
    elif rating == '20+ ratings':
        return randint(21, 50)
    elif rating == '50+ ratings':
        return randint(51, 100)
    elif rating == '100+ ratings':
        return randint(101, 200)
    elif rating == '500+ ratings':
        return randint(501, 1000)
    elif rating == '1K+ ratings':
        return randint(1001, 5000)
    elif rating == '5K+ ratings':
        return randint(5001, 10000)
    elif rating == '10K+ ratings':
        return randint(10001, 20000)
    else:
        return 0

dfSwiggy['costvalue']=dfSwiggy['cost'].str.replace('₹', '').str.replace(' ', '').astype(float)
dfSwiggy['ratingcount']=dfSwiggy['rating_count'].apply(convert_ratings).astype(int)
dfSwiggy.head()
```

```python
dfSwiggy.info()
```

```python
dfSwiggy.isnull().sum()
```

```python
dfSwiggy.dropna(subset=['name'], inplace=True)
dfSwiggy['rating'] = dfSwiggy['rating'].fillna(0)

dfSwiggy['cost'] = dfSwiggy['cost'].fillna(dfSwiggy['cost'].mode()[0])
dfSwiggy['cuisine'] = dfSwiggy['cuisine'].fillna(dfSwiggy['cuisine'].mode()[0])
dfSwiggy['costvalue'] = dfSwiggy['costvalue'].fillna(dfSwiggy['costvalue'].mode()[0]) 

dfSwiggy.isnull().sum()
```

```python
dfSwiggy.info()
dfSwiggy.describe()
```

```python
def split_column(value):
    parts = value.split(',')
    if len(parts) > 1:
        return parts[0], ','.join(parts[1:])
    else:
        return parts[0], ''

dfSwiggy[['Place', 'Maincity']] = dfSwiggy['city'].apply(lambda x: pd.Series(split_column(x)))
dfSwiggy.head()
```

```python
dfSwiggy.info()
dfSwiggy['Maincity'].unique()
```
