# 2. Data Preprocessing:

```python
import pickle
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_cols = ['name', 'Place', 'Maincity', 'cuisine']
top_n = 50

for col in categorical_cols:
    top_categories = dfSwiggy[col].value_counts().nlargest(top_n).index
    dfSwiggy[col] = dfSwiggy[col].apply(lambda x: x if x in top_categories else "Other")

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoded_data = encoder.fit_transform(dfSwiggy[categorical_cols])

dfEncoded = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

with open('encoder.pkl', 'wb') as f:
    pickle.dump(encoder, f)
```

```python
dfencodeddata = pd.concat([dfSwiggy.loc[:,['costvalue','rating']], dfEncoded], axis=1)
dfencodeddata.head()
```
